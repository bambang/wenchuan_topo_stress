```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import halfspace.projections as hsp
from matplotlib import gridspec
```

```python
zdf = pd.read_csv('../results/zhang_fail_posteriors.csv', index_col=0)
```

```python
zdf.head()
```

```python
fdf = pd.read_csv('../results/feng_fail_posteriors.csv', index_col=0)
qrdf = pd.read_csv('../results/qi_rough_fail_posteriors.csv', index_col=0)
qsdf = pd.read_csv('../results/qi_smooth_fail_posteriors.csv', index_col=0)
```

```python
fdf.head()
```

```python
all_inds = list( set(zdf.index.values) & set(fdf.index.values) 
                 & set(qrdf.index.values) & set(qsdf.index.values) )
```

```python
zdf.loc[all_inds].head()
```

```python
def get_stress_comps(df):
    XX = []
    xx = []
    YY = []
    yy = []

    for i in range(len(df.txx)):
        X1, Y1, x1, y1 = hsp.calc_xy_princ_stresses_from_stress_comps(
                           df.txx.values[i], df.tyy.values[i], df.txy.values[i])

        XX.append(X1)
        YY.append(Y1)
        xx.append(x1)
        yy.append(y1)

    XX = np.array(XX)
    YY = np.array(YY)
    xx = np.array(xx)
    yy = np.array(yy)
    
    return XX, YY, xx, yy



def plot_stresses(XX, YY, xx, yy, model='None', m_size=2, alph=0.1, circ_color='grey', circ_lw=0.5):
    

    plt.figure(1)

    plt.subplot(1,3,1)

    rad_1 = plt.Circle((0,0), 1, color=circ_color, fill=False, lw=circ_lw)
    plt.gca().add_artist(rad_1)

    plt.scatter(XX, YY, c='r', s=m_size, lw=0, alpha=alph)
    plt.scatter(-XX, -YY, c='r', s=m_size, lw=0, alpha=alph)
    plt.axis('equal')


    plt.subplot(1,3,2)
    
        
    plt.title(model)
    
    rad_1 = plt.Circle((0,0), 1, color=circ_color, fill=False, lw=circ_lw)
    plt.gca().add_artist(rad_1)

    plt.scatter(xx, yy, c='b', s=m_size, lw=0, alpha=alph)
    plt.scatter(-xx, -yy, c='b', s=m_size, lw=0, alpha=alph)
    plt.axis('equal')


    plt.subplot(1,3,3)

    rad_1 = plt.Circle((0,0), 1, color=circ_color, fill=False, lw=circ_lw)
    plt.gca().add_artist(rad_1)

    plt.scatter(XX, YY, c='r', s=m_size, lw=0, alpha=alph)
    plt.scatter(-XX, -YY, c='r', s=m_size, lw=0, alpha=alph)

    plt.scatter(xx, yy, c='b', s=m_size, lw=0, alpha=alph)
    plt.scatter(-xx, -yy, c='b', s=m_size, lw=0, alpha=alph)

    plt.axis('equal')
    
    #return fig

def scatter_w_marginals(x_data, y_data, c=None,
                        lw=1, s=10, alpha=1,
                        bins=10, xlabel='',
                        ylabel='', x_lim=None, y_lim=None):
    
    figure()
    if not x_lim:
        x_lim = [x_data.min(), x_data.max()]
    if not ylim:
        y_lim = [y_data.min(), y_data.max()]
    
    gs = gridspec.GridSpec(2, 2, width_ratios=[3,1],
                                 height_ratios=[1,3])
    ax = subplot(gs[1,0])
    axt = subplot(gs[0,0], sharex=ax)
    axl = subplot(gs[1,1], sharey=ax)
    
    axt.hist(x_data, bins=bins)
    axl.hist(y_data, bins=bins, orientation='horizontal')
    
    sc = ax.scatter(x_data, y_data, c=c, lw=lw, s=s, alpha=alpha)
    #colorbar(sc)
    ax.set_xlabel(xlabel)
    ax.set_ylabel(ylabel)
    #ax.colorbar()
    #ax.set_xlim([0,1])
    #ax.set_ylim([0,1])
    ax.set_xlim( x_lim )
    ax.set_ylim( y_lim )
```

```python
zXX, zYY, zxx, zyy = get_stress_comps(zdf)
```

```python
fXX, fYY, fxx, fyy = get_stress_comps(fdf)
qrXX, qrYY, qrxx, qryy = get_stress_comps(qrdf)
qsXX, qsYY, qsxx, qsyy = get_stress_comps(qsdf)
```

```python
adf = zdf.loc[all_inds, ['txx', 'tyy', 'txy', 'phi']]

aXX, aYY, axx, ayy = get_stress_comps(adf)
```

```python
f_t_max = np.sqrt( fXX**2 + fYY**2)
z_t_max = np.sqrt( zXX**2 + zYY**2)
qr_t_max = np.sqrt( qrXX**2 + qrYY**2)
qs_t_max = np.sqrt( qsXX**2 + qsYY**2)
a_t_max = np.sqrt( aXX**2 + aYY**2)

f_t_min = np.sqrt( fxx**2 + fyy**2)
z_t_min = np.sqrt( zxx**2 + zyy**2)
qr_t_min = np.sqrt( qrxx**2 + qryy**2)
qs_t_min = np.sqrt( qsxx**2 + qsyy**2)
a_t_min = np.sqrt( axx**2 + ayy**2)
```

```python
adf['mu_mean'] = ( zdf.loc[all_inds, 'mu'] + fdf.loc[all_inds, 'mu'] + 
                   qrdf.loc[all_inds, 'mu'] + qsdf.loc[all_inds, 'mu'] ) / 4.
```

```python
plot_stresses(zXX, zYY, zxx, zyy, model='zhang', alph=0.05)

fig = scatter_w_marginals(z_t_max, z_t_min, lw=0, c=zdf.txx, bins=50,
                          xlabel='T max', ylabel='T min')


fig = scatter_w_marginals(zdf.phi.values, zdf.mu.values, lw=0, c=z_t_max, bins=50,
                          xlabel='fluid pressure', ylabel='friction', x_lim=[0,1],
                          y_lim=[0,1])
```

```python
plot_stresses(fXX, fYY, fxx, fyy, model='feng', alph=0.05)

fig = scatter_w_marginals(f_t_max, f_t_min, lw=0, c=fdf.txx, bins=50,
                          xlabel='T max', ylabel='T min')


fig = scatter_w_marginals(fdf.phi.values, fdf.mu.values, lw=0, c=f_t_max, bins=50,
                          xlabel='fluid pressure', ylabel='friction', x_lim=[0,1],
                          y_lim=[0,1])
```

```python
plot_stresses(qrXX, qrYY, qrxx, qryy, model='qi rough', alph=0.05)

fig = scatter_w_marginals(qr_t_max, qr_t_min, lw=0, c=qrdf.txx, bins=50,
                          xlabel='T max', ylabel='T min')


fig = scatter_w_marginals(qrdf.phi.values, qrdf.mu.values, lw=0, c=qr_t_max, bins=50,
                          xlabel='fluid pressure', ylabel='friction', x_lim=[0,1],
                          y_lim=[0,1])
```

```python
plot_stresses(qsXX, qsYY, qsxx, qsyy, model='qi smooth', alph=0.05)

fig = scatter_w_marginals(qs_t_max, qs_t_min, lw=0, c=qsdf.txx, bins=50,
                          xlabel='T max', ylabel='T min')


fig = scatter_w_marginals(qsdf.phi.values, qsdf.mu.values, lw=0, c=qs_t_max, bins=50,
                          xlabel='fluid pressure', ylabel='friction', x_lim=[0,1],
                          y_lim=[0,1])
```

```python
plot_stresses(aXX, aYY, axx, ayy, model='all', alph=0.05)

fig = scatter_w_marginals(a_t_max, a_t_min, lw=0, c=adf.txx, bins=50,
                          xlabel='T max', ylabel='T min')


fig = scatter_w_marginals(adf.phi.values, adf.mu_mean.values, lw=0, c=a_t_max, bins=50,
                          xlabel='fluid pressure', ylabel='friction', x_lim=[0,1],
                          y_lim=[0,1])
```
