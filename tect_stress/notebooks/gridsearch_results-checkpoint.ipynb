```python
import numpy as np
import pandas as pd
```

```python
search_df = pd.read_csv('tect_search_df.csv', index_col=0)
```

```python
iters = search_df.groupby('iter')
```

```python
iters.describe()
```

```python
means = iters[['tect_xx', 'tect_yy', 'tect_xy', 'slip_rake', 'tau_rake',
               'misfit', 'weighted_misfit']].mean()
```

```python
means
```

```python
plot(iters.misfit.mean(),iters.misfit.std(), ',', alpha=1)
```

```python
plot(iters.weighted_misfit.mean(),iters.weighted_misfit.std(), ',', alpha=1)
```

```python
total_slip = 2.358513 * 702
```

```python
plot(iters.weighted_misfit.mean(), 
     iters.weighted_misfit.std() / total_slip,
     ',', alpha=1)
```

```python
plot(iters.misfit.mean(), iters.weighted_misfit.mean(), ',')
xlabel('mean misfit')
ylabel('mean weighted misfit')
```

```python
iters.weighted_misfit.mean().hist(bins=100)
```

```python
means.weighted_misfit.hist(bins=100)
```

```python
w_good = means[(means.weighted_misfit <= 20) & (means.weighted_misfit >=-20)]
w_best = means[(means.weighted_misfit <= 5) & (means.weighted_misfit >=-5)]
```

```python
w_good.weighted_misfit.hist()
```

```python
scatter(w_good.tect_xx, w_good.tect_yy, s=w_good.tect_xy,
        c=w_good.weighted_misfit, lw=0)

colorbar(label='weighted misfit')
xlabel('E stress')
ylabel('N stress')

show()
```

```python
plot(w_good.weighted_misfit, w_good.misfit, '.')
```

```python
scatter(w_best.tect_xx, w_best.tect_yy, s=w_best.tect_xy,
        c=w_best.weighted_misfit, lw=0)

colorbar(label='weighted misfit')
xlabel('E stress')
ylabel('N stress')

show()
```
