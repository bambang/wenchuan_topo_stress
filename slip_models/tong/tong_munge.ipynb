```python
t2x_c = np.mean( t2[t2['y1']==-0][['x1','x3']].values )
t3x_c = np.mean( t3[t3['y1']==-0][['x1','x3']].values )
t4x_c = np.mean( t4[t4['y1']==-0][['x1','x3']].values )
```

```python
t1['xc'] = (t1.x1 + t1.x3)/2.
t2['xc'] = (t2.x1 + t2.x3)/2.
t3['xc'] = (t3.x1 + t3.x3)/2.
t4['xc'] = (t4.x1 + t4.x3)/2.
```

```python
t1['xc_'] = (t1.x1 + t1.x3)/2. - t1x_c
t2['xc_'] = (t2.x1 + t2.x3)/2. - t2x_c
t3['xc_'] = (t3.x1 + t3.x3)/2. - t3x_c
t4['xc_'] = (t4.x1 + t4.x3)/2. - t4x_c
```

```python
t1['yc'] = (t1.y1 + t1.y3)/2.
t3['yc'] = (t3.y1 + t3.y3)/2.
t2['yc'] = (t2.y1 + t2.y3)/2.
t4['yc'] = (t4.y1 + t4.y3)/2.
```

```python
t1['yc_'] = t1.yc * np.cos( np.radians( t1p['dip']) )
t1['z_'] = t1.yc * np.sin( np.radians( t1p['dip']) )
t2['yc_'] = t2.yc * np.cos( np.radians( t2p['dip']) )
t2['z_'] = t2.yc * np.sin( np.radians( t2p['dip']) )
t3['yc_'] = t3.yc * np.cos( np.radians( t3p['dip']) )
t3['z_'] = t3.yc * np.sin( np.radians( t3p['dip']) )
t4['yc_'] = t4.yc * np.cos( np.radians( t4p['dip']) )
t4['z_'] = t4.yc * np.sin( np.radians( t4p['dip']) )
```

```python
t1['xr_'], t1['yr_'] = hsp.rotate_pts_2d(t1.xc_.values, t1.yc_.values,
                                         rotation_angle=t1d['strike'])

scatter(t1.xr_, t1.yr_, c=t1.z_)
axis('equal')
show()
```

### Good... now let's put these in real coordinates.

```python
wgs84 = pj.Proj(init='epsg:4326')
utm48 = pj.Proj(init='epsg:32648')
```

```python
east_origin, north_origin = pj.transform(wgs84, utm48, 
                                         lon_origin, lat_origin)
```

```python
t1['east_utm48'] = np.int_(np.round( t1.xg + east_origin) )
t2['east_utm48'] = np.int_(np.round( t2.xg + east_origin) )
t3['east_utm48'] = np.int_(np.round( t3.xg + east_origin) )
t4['east_utm48'] = np.int_(np.round( t4.xg + east_origin) )

t1['north_utm48'] = np.int_(np.round( t1.yg + north_origin) )
t2['north_utm48'] = np.int_(np.round( t2.yg + north_origin) )
t3['north_utm48'] = np.int_(np.round( t3.yg + north_origin) )
t4['north_utm48'] = np.int_(np.round( t4.yg + north_origin) )
```

```python
t1['segment']=1
t2['segment']=2
t3['segment']=3
t4['segment']=4
```

```python
t1['strike'] = t1p['strike']
t2['strike'] = t2p['strike']
t3['strike'] = t3p['strike']
t4['strike'] = t4p['strike']

t1['dip'] = t1p['dip']
t2['dip'] = t3p['dip']
t3['dip'] = t3p['dip']
t4['dip'] = t4p['dip']
```

```python
t1['net_slip'] = (t1.strike_slip**2 + t1.dip_slip**2)**0.5
t2['net_slip'] = (t2.strike_slip**2 + t2.dip_slip**2)**0.5
t3['net_slip'] = (t3.strike_slip**2 + t3.dip_slip**2)**0.5
t4['net_slip'] = (t4.strike_slip**2 + t4.dip_slip**2)**0.5
```

```python
save_cols = ['east_utm48', 'north_utm48', 'z', 'strike_slip',
             'dip_slip', 'net_slip', 'strike', 'dip']
```
