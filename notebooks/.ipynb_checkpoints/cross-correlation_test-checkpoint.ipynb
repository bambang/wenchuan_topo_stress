extrapolate faults:

1) do vertical axis rotation, so along_strike is x=0, sw (or whatever) point is origin

2) extend fault in both directions (y = +/- whatever, x = 0, z = [0-max depth])

3) calculate x coordinate as a function of dip

4) vertical axis rotate back

###look at cross-correlation

make arrays

```python
f3_depth_vec = f3.depth.unique()
```

```python
f3_depth_vec
```

```python
f3_along_strike = f3[f3.depth == f3.depth.min()]['yr'].values
```

```python
f3_along_strike = f3_along_strike[::-1]
```

```python
f3_yy, f3_zz = np.meshgrid(f3_along_strike, f3_depth_vec)
```

```python
d_slip_grid = sci.griddata( (f3.yr, f3.depth), f3.dip_m, (f3_yy, f3_zz), 
                           method='linear', fill_value=0)

s_slip_grid = sci.griddata( (f3.yr, f3.depth), -f3.strike_m, (f3_yy, f3_zz), 
                           method='linear', fill_value=0)
```

```python
pcolormesh(d_slip_grid)
colorbar()
```

```python
pcolormesh(s_slip_grid)
colorbar()
```

```python
%%timeit

scs.correlate2d(s_slip_grid, d_slip_grid)
```

```python
pcolormesh( scs.correlate2d(s_slip_grid, d_slip_grid, mode='same') )
colorbar()
plot( s_slip_grid.shape[1]/2, s_slip_grid.shape[0]/2, 'yo')
#plot( 40, 8, 'yo')
```

```python
d_tau_grid = sci.griddata( (f3.yr, f3.depth), f3.tau_dd, (f3_yy, f3_zz), 
                           method='linear', fill_value=0)

s_tau_grid = sci.griddata( (f3.yr, f3.depth), -f3.tau_ss, (f3_yy, f3_zz), 
                           method='linear', fill_value=0)

n_sig_grid = sci.griddata( (f3.yr, f3.depth), f3.sig_nn, (f3_yy, f3_zz), 
                           method='linear', fill_value=0)
```

```python
cc = scs.correlate2d(d_slip_grid, d_tau_grid, mode='full')

pcolormesh( cc )
colorbar()
plot( cc.shape[1]/2, cc.shape[0]/2, 'yo')
```

```python
cc = scs.correlate2d( zscore(d_slip_grid, axis=None), zscore(d_tau_grid, axis=None), mode='full')

pcolormesh( cc )
colorbar()
plot( cc.shape[1]/2, cc.shape[0]/2, 'yo')
```
