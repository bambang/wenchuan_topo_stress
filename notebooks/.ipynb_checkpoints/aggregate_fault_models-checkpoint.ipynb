```python
zhang_lms = pd.read_csv('zhang/lms_stress_slip.csv', index_col=0)
```

```python
zhang_lms.head()
```

```python
plot(zhang_lms.lon_deg, zhang_lms.lat_deg, '.')
```

```python
import pyproj as pj
```

```python
wgs84 = pj.Proj(init='epsg:4326')
utm48 = pj.Proj(init='epsg:32648')
```

```python
zhang_lms['easting_utm48'], zhang_lms['northing_utm48'] = pj.transform(
                                    wgs84, utm48, zhang_lms.lon_deg.values,
                                    zhang_lms.lat_deg.values)
```

```python
zhang_lms.dip_deg.unique()
```

```python
zhang_peng = zhang_lms[zhang_lms.dip_deg == 33.].copy()
```

```python
zhang_beich = zhang_lms[zhang_lms.dip_deg != 33.]
```

```python
zhang_lms.dip_deg.unique()
```

```python
plot(zhang_beich.easting_utm48, zhang_beich.northing_utm48, '.')
```

```python
zhang_beich.slp_ddip_m.hist()
```

```python
zhang_beich.slp_strk_m.hist()
```

```python
zhang_beich.to_csv('zhang_beich.csv')
zhang_peng.to_csv('zhang_peng.csv')
```
