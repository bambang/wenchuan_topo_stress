extrapolate faults:

1) do vertical axis rotation, so along_strike is x=0, sw (or whatever) point is origin

2) extend fault in both directions (y = +/- whatever, x = 0, z = [0-max depth])

3) calculate x coordinate as a function of dip

4) vertical axis rotate back

add/subtract unrotated utm coord (diffs)

```python
f3_ne = {}
f3_sw = {}
```

```python
f3_ne['utm_lon'] = f3.utm_lon.copy() + (f3[f3.depth == f3.depth.min()].utm_lon.max() 
                                     - f3[f3.depth == f3.depth.min()].utm_lon.min() )

f3_ne['utm_lat'] = f3.utm_lat.copy() + (f3[f3.depth == f3.depth.min()].utm_lat.max() 
                                     - f3[f3.depth == f3.depth.min()].utm_lat.min() )

f3_ne['depth'] = f3.depth.copy()


f3_sw['utm_lon'] = f3.utm_lon.copy() - (f3[f3.depth == f3.depth.min()].utm_lon.max() 
                                     - f3[f3.depth == f3.depth.min()].utm_lon.min() )

f3_sw['utm_lat'] = f3.utm_lat.copy() - (f3[f3.depth == f3.depth.min()].utm_lat.max() 
                                     - f3[f3.depth == f3.depth.min()].utm_lat.min() )

f3_sw['depth'] = f3.depth.copy()
```

```python
f3_ne = pd.DataFrame(data = f3_ne)
f3_sw = pd.DataFrame(data = f3_sw)
```

```python
scatter(f3.utm_lon, f3.utm_lat, c=f3.depth, s=30, lw=0)
scatter(f3_ne.utm_lon, f3_ne.utm_lat, c='r', s=5, lw=0)
scatter(f3_sw.utm_lon, f3_sw.utm_lat, c='g', s=5, lw=0)

axis('equal')
show()
```

```python
f3_c_coords = f3[['utm_lon', 'utm_lat', 'depth']]
```

```python
f3_long_shallow = pd.concat([f3_c_coords, f3_ne, f3_sw], axis=0, ignore_index=True)
```

```python
f3_long_shallow
```

```python
f3_deep = f3_long_shallow.copy()

f3_deep.utm_lon += ( f3_long_shallow[f3_long_shallow.depth == 
                                    f3_long_shallow.depth.max()].utm_lon.min()
                    -f3_long_shallow[f3_long_shallow.depth == 
                                    f3_long_shallow.depth.min()].utm_lon.min() )
                    
f3_deep.utm_lat += ( f3_long_shallow[f3_long_shallow.depth == 
                                    f3_long_shallow.depth.max()].utm_lat.min()
                    -f3_long_shallow[f3_long_shallow.depth == 
                                    f3_long_shallow.depth.min()].utm_lat.min() )

f3_deep.depth += f3_long_shallow.depth.max()
```

```python
f3_up = f3_long_shallow.copy()

f3_up.utm_lon += -( f3_long_shallow[f3_long_shallow.depth == 
                                    f3_long_shallow.depth.max()].utm_lon.min()
                    -f3_long_shallow[f3_long_shallow.depth == 
                                    f3_long_shallow.depth.min()].utm_lon.min() )
                    
f3_up.utm_lat += -( f3_long_shallow[f3_long_shallow.depth == 
                                    f3_long_shallow.depth.max()].utm_lat.min()
                    -f3_long_shallow[f3_long_shallow.depth == 
                                    f3_long_shallow.depth.min()].utm_lat.min() )

f3_up.depth += -f3_long_shallow.depth.max()
```

```python
d_tau_grid = sci.griddata( (f3.yr, f3.depth), f3.tau_dd, (f3_yy, f3_zz), 
                           method='linear', fill_value=0)

s_tau_grid = sci.griddata( (f3.yr, f3.depth), -f3.tau_ss, (f3_yy, f3_zz), 
                           method='linear', fill_value=0)

n_sig_grid = sci.griddata( (f3.yr, f3.depth), f3.sig_nn, (f3_yy, f3_zz), 
                           method='linear', fill_value=0)
```
