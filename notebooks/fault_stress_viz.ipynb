```python
zz = pd.read_csv('../slip_models/zhang/lms_stress_slip.csv', index_col=0)

# shallowest level is above top of halfspace; trim here
zz = zz[zz.depth_km > 1.4993]
```

```python
zz['S_tau_s'] = scv.strike_shear( strike=zz.strike_deg, dip=zz.dip_deg, 
                                 rho = rho, g = g, mxx = zz.xx_stress*1e6,
                                 myy = zz.yy_stress*1e6, mxy=zz.xy_stress*1e6,
                                 mzz=zz.zz_stress*1e6, mxz=zz.xz_stress*1e6,
                                 myz=zz.yz_stress*1e6, txx=hd['txx'], tyy=hd['tyy'], txy=hd['txy'],
                                 depth=zz.depth_km*1000) * 1e-6

zz['S_tau_d'] = scv.dip_shear( strike=zz.strike_deg, dip=zz.dip_deg, 
                                 rho = rho, g = g, mxx = zz.xx_stress*1e6,
                                 myy = zz.yy_stress*1e6, mxy=zz.xy_stress*1e6,
                                 mzz=zz.zz_stress*1e6, mxz=zz.xz_stress*1e6,
                                 myz=zz.yz_stress*1e6, txx=hd['txx'], tyy=hd['tyy'], txy=hd['txy'],
                                 depth=zz.depth_km*1000) * 1e-6

zz['sig_n_eff'] = scv.eff_normal_stress( strike=zz.strike_deg, dip=zz.dip_deg, 
                                 rho = rho, g = g, mxx = zz.xx_stress*1e6,
                                 myy = zz.yy_stress*1e6, mxy=zz.xy_stress*1e6,
                                 mzz=zz.zz_stress*1e6, mxz=zz.xz_stress*1e6,
                                 myz=zz.yz_stress*1e6, txx=hd['txx'], tyy=hd['tyy'], txy=hd['txy'],
                                 depth=zz.depth_km*1000, phi=hd['phi'])  * 1e-6

zz['S_tau_mag'] = np.sqrt(zz.S_tau_s**2 + zz.S_tau_d**2)

zz['S_rake'] = hsp.get_rake_from_shear_components(strike_shear=zz.S_tau_s, dip_shear=zz.S_tau_d)
```

```python
zz['cfs'] = zz.S_tau_mag - hd['mu'] * zz.sig_n_eff
```
