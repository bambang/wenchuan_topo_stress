## Some descriptive rake statistics

```python
import halfspace.projections as hsp
```

```python
bf.iloc[0]
```

```python
bf['qa_rake'] = hsp.get_rake_from_shear_components(bf.qa_strike_slip,
                                                    bf.qa_dip_slip)

bf['z_rake'] = hsp.get_rake_from_shear_components(bf.z_strike_slip,
                                                    bf.z_dip_slip)

bf['f_rake'] = hsp.get_rake_from_shear_components(bf.f_strike_slip,
                                                    bf.f_dip_slip)
```

```python
bf['qa_slip_mag'] = np.sqrt(bf.qa_strike_slip**2 + bf.qa_dip_slip**2)
bf['z_slip_mag'] = np.sqrt(bf.z_strike_slip**2 + bf.z_dip_slip**2)
bf['f_slip_mag'] = np.sqrt(bf.f_strike_slip**2 + bf.f_dip_slip**2)
```

```python
scatter(bf.qa_slip_mag, bf.qa_rake, c='r', alpha=0.5, lw=0)
scatter(bf.z_slip_mag, bf.z_rake, c='b', alpha=0.5, lw=0)
scatter(bf.f_slip_mag, bf.f_rake, c='g', alpha=0.5, lw=0)

ylim([-10, 100])

show()
```

```python
scatter(bf.qa_strike_slip, bf.qa_dip_slip, c='r', alpha=0.5, lw=0)
scatter(bf.z_strike_slip, bf.z_dip_slip, c='b', alpha=0.5, lw=0)
scatter(bf.f_strike_slip, bf.f_dip_slip, c='g', alpha=0.5, lw=0)
```

```python
scatter(bf.slip_mag_mean, np.std([bf.qa_rake, bf.z_rake, bf.f_rake], axis=0))

xlabel('mean slip (m)')
ylabel('rake st. dev (deg)')
```

```python
scatter(bf.slip_mag_med, sm.robust.scale.stand_mad([bf.qa_rake, bf.z_rake, 
                                                    bf.f_rake], axis=0))

xlabel('mean slip (m)')
ylabel('rake st. dev (deg)')
```
