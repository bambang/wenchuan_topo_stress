```python
zdf_relevant_cols = zdf[['depth_km', 'strike_deg', 'dip_deg', 'xx_stress', 'yy_stress', 'zz_stress',
          'xz_stress', 'yz_stress', 'xy_stress']]
```

```python
rho = 2700
g = 9.81

litho_z = rho * g / np.float(1e3)
```

make dataframe with only relevant columns

```python
zdf_relevant_cols = zdf_relevant_cols.reset_index()
```

populate dataframe with results randomly drawn from the joint posteriors,
for each point in the fault model

```python
n_samps = 1000
```

```python
idx_samp = np.random.choice(pTD.index, n_samps)
```

```python
mc_zdf = pd.DataFrame(np.tile(zdf_relevant_cols.values, [n_samps,1]), 
                      columns=zdf_relevant_cols.columns)
```

```python
mc_zdf.columns = [u'pt_index', u'depth_km', u'strike_deg', u'dip_deg', 
                  u'mxx', u'myy', u'mzz', u'mxz', 
                  u'myz', u'mxy']
```

```python
mc_zdf['iter'] = 0
```

```python
nstart = 0

for samp in np.arange(n_samps):
    
    nstop = nstart + len(zdf_relevant_cols.index)
    
    mc_zdf.loc[nstart:nstop, 'iter'] = samp
    
    nstart += len(zdf_relevant_cols.index)
```

```python
mc_zdf['mu']  = 0.
mc_zdf['phi'] = 0.
mc_zdf['txx'] = 0.
mc_zdf['tyy'] = 0.
mc_zdf['txy'] = 0.

nstart = 0

for samp in np.arange(n_samps):

    nstop = nstart + len(zdf_relevant_cols.index)
    
    mc_zdf.loc[nstart:nstop, ['txx', 'tyy', 'txy', 'mu', 'phi']] = (
                        pTD.loc[ idx_samp[samp]].values)
    
    nstart += len(zdf_relevant_cols.index)
```

```python
def make_S(row):
    S = hsp.make_xyz_stress_tensor(
        sig_xx= row['mxx'] + ( (row['txx'] + 1)* row['depth_km'] * litho_z),
        sig_yy= row['myy'] + ( (row['tyy'] + 1)* row['depth_km'] * litho_z),
        sig_xy= row['mxy'] + (  row['txy'] * row['depth_km'] * litho_z),
        sig_zz= row['mzz'] + ( row['depth_km'] * litho_z),
        sig_xz= row['mxz'], sig_yz= row['myz']
        )
    
    return S

def get_tau_f(row):
    
    S = make_S(row)
    
    return hsp.max_shear_stress_from_xyz(strike=row['strike_deg'],
                                         dip=row['dip_deg'], 
                                         stress_tensor=S)[0]

def get_tau_o(row):
    S = make_S(row)
    
    return hsp.shear_stress_on_optimal_plane(S, friction_coefficent=0.6)[0]
    

def get_sig_n_f(row):
    S = make_S(row)
    
    sig_n = hsp.normal_stress_from_xyz(strike=row['strike_deg'],
                                         dip=row['dip_deg'], 
                                         stress_tensor=S)
    return sig_n

def get_sig_n_o(row):
    S = make_S(row)
    
    return hsp.normal_stress_on_optimal_plane(S, friction_coefficient=0.6)

def get_opt_s_d(row):
    S = make_S(row)
    
    s, d = hsp.find_optimal_plane(S, friction_coefficient=0.6)
    
    return pd.Series({'opt_strike':s, 'opt_dip':d})
```

```python
t0 = time.time()

mc_zdf['tau_f'] = mc_zdf.apply(get_tau_f, axis=1)

print(time.time() - t0)
```

```python
t0 = time.time()

mc_zdf['tau_o'] = mc_zdf.apply(get_tau_o, axis=1)

print(time.time() - t0)
```

```python
t0 = time.time()

mc_zdf['sig_n_f'] = mc_zdf.apply(get_sig_n_f, axis=1)

print(time.time() - t0)
```

```python
t0 = time.time()

mc_zdf['sig_n_o'] = mc_zdf.apply(get_sig_n_o, axis=1)

print(time.time() - t0)
```

```python
mc_zdf[['opt_dip', 'opt_strike']] = mc_zdf.apply(get_opt_s_d, axis=1)
```

```python
mc_zdf['CFR_f'] = mc_zdf.tau_f / (mc_zdf.mu * (1 - mc_zdf.phi) * mc_zdf.sig_n_f)
mc_zdf['CFR_o'] = mc_zdf.tau_o / (0.6 * (1 - mc_zdf.phi) * mc_zdf.sig_n_o)
```

```python
mc_zdf['cff_ratio'] = mc_zdf['CFR_o'] / mc_zdf['CFR_f']

mc_zdf['log_cff_ratio'] = np.log(mc_zdf.cff_ratio)
```

```python
def calc_xy_max_stress_from_stress_comps(s_xx=0., s_yy=0., s_xy=0.):
    '''
    Takes the three indepenent stress tensor components for a 2 dimensional
    stress tensor and returns the magnitude and azimuth of the maximum
    principal stress.

    Returns a tuple: max_magnitude, max_azimuth
    '''
    s_xx, s_yy, s_xy = row['txx'], row['tyy'], row['txy']
    
    max_x, max_y, min_x, min_y = hsp.calc_xy_princ_stresses_from_stress_comps(
                                        s_xx=s_xx, x_yy=s_yy, s_xy=s_xy)

    max_mag = np.sqrt(max_x**2 + max_y**2)

    max_az = hsp.angle_to_azimuth( np.arctan2(max_y, max_x) )

    return max_mag, max_az
```
