```python
t1['xc'] = (t1.x1 + t1.x3)/2.
t1['yc'] = (t1.y1 + t1.y3)/2.
```

```python
plot(t1.xc.values, t1.yc.values, '.')
```

```python
t2['xc'] = (t2.x1 + t2.x3)/2.
t2['yc'] = (t2.y1 + t2.y3)/2.
t3['xc'] = (t3.x1 + t3.x3)/2.
t3['yc'] = (t3.y1 + t3.y3)/2.
t4['xc'] = (t4.x1 + t4.x3)/2.
t4['yc'] = (t4.y1 + t4.y3)/2.
```

```python
plot(t1.xc.values, t1.yc.values, '.')
plot(t2.xc.values, t2.yc.values, '.')
plot(t3.xc.values, t3.yc.values, '.')
axis('equal')
```

```python
t1['y_'] = t1.yc * np.cos( np.radians( t1p['dip']) )
t1['z_'] = t1.yc * np.sin( np.radians( t1p['dip']) )
```

```python
plot(t1.y_.values, t1.z_.values, '.')
axis('equal')
```

```python
t2['y_'] = t2.yc * np.cos( np.radians( t2p['dip']) )
t2['z_'] = t2.yc * np.sin( np.radians( t2p['dip']) )
t3['y_'] = t3.yc * np.cos( np.radians( t3p['dip']) )
t3['z_'] = t3.yc * np.sin( np.radians( t3p['dip']) )
t4['y_'] = t4.yc * np.cos( np.radians( t4p['dip']) )
t4['z_'] = t4.yc * np.sin( np.radians( t4p['dip']) )
```

```python
plot(t1.xc.values, t1.z_.values, '.')
plot(t2.xc.values, t2.z_.values, '.')
plot(t3.xc.values, t3.z_.values, '.')
axis('equal')
```

```python
plot(t1.y_.values, t1.z_.values, '.')
plot(t2.y_.values, t2.z_.values, '.')
plot(t3.y_.values, t3.z_.values, '.')
#plot(t1.y_.values, t1.z_.values, '.')
axis('equal')
```

```python
plot(t1.xc.values, t1.y_.values, 'r.')
plot(t2.xc.values, t2.y_.values, 'g.')
plot(t3.xc.values, t3.y_.values, 'b.')
```

```python
t1['xr'], t1['yr'] = hsp.rotate_pts_2d(t1.xc, t1.y_, 
                                       rotation_angle=t1p['strike'],
                                       angle_input='degrees')
```

```python
t1['xr'], t1['yr'] = hsp.rotate_pts_2d(t1.xc, t1.y_, 
                                       x0 = -t1p['xo'], y0 = -t1p['yo'],
                                       rotation_angle = -t1p['strike']+180,
                                       angle_input='degrees')

t2['xr'], t2['yr'] = hsp.rotate_pts_2d(t2.xc, t2.y_, 
                                       x0 = -t2p['xo'], y0 = -t2p['yo'],
                                       rotation_angle = -t2p['strike']+180,
                                       angle_input='degrees')

t3['xr'], t3['yr'] = hsp.rotate_pts_2d(t3.xc, t3.y_, 
                                       x0 = -t3p['xo'], y0 = -t3p['yo'],
                                       rotation_angle = -t3p['strike']+180,
                                       angle_input='degrees')

t4['xr'], t4['yr'] = hsp.rotate_pts_2d(t4.xc, t4.y_, 
                                       x0 = -t4p['xo'], y0 = -t4p['yo'],
                                       rotation_angle = -t4p['strike']+180,
                                       angle_input='degrees')
```
